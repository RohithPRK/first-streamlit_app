```python
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task="table-question-answering",model="google/tapas-base-finetuned-wtq")
```

```python
table=pd.read_csv("/content/Urovant - Tableau (1).csv")
```

```python
table = table.astype(str)
```

```python
table
```

```python
query = "Give me the flag names with PDRP 'YES'"
print(tqa(table=table,query=query)["answer"])

```

```python
query = "list the trxeq and nrxeq values"
print(tqa(table=table,query=query)["answer"])
```

```python
query = "what is the flag for smallest NRXEQ"
print(tqa(table=table,query=query)["answer"])
```
